# Q1. Explain GET and POST methods.


## GET Method:
- Used to request data from the server. Safe and idempotent.

## POST Method:
- Used to submit data to the server. Not safe, may cause side effects.

# Q2. Why is request used in Flask?


* The "request" object in Flask is used to access and process data sent by the client in an incoming HTTP request. 
* It allows Flask applications to handle form data, retrieve query parameters, access headers and cookies, and more. 
* It's essential for building interactive web applications.

# Q3. Why is redirect() used in Flask?

redirect() in Flask is used to send a redirection response to the client's web browser, guiding it to a different URL or route within the application. It's commonly used for handling form submissions, dynamic redirection, and creating user-friendly URLs.

# Q4. What are templates in Flask? Why is the render_template() function used?


Templates in Flask are files that separate presentation (HTML) from application logic. They allow dynamic content rendering. render_template() is used to combine data with templates, generating HTML responses to send back to clients. It promotes code organization and creates interactive web pages.

# Q5. Create a simple API. Use Postman to test it. Attach the screenshot of the output in the Jupyter Notebook

```python
from flask import Flask, jsonify

app = Flask(__name__)

# Sample data - list of fruits
fruits = [
    {"id": 1, "name": "Apple", "color": "Red"},
    {"id": 2, "name": "Banana", "color": "Yellow"},
    {"id": 3, "name": "Orange", "color": "Orange"}
]

# Route to get all fruits
@app.route('/api/fruits', methods=['GET'])
def get_fruits():
    return jsonify(fruits)

if __name__ == '__main__':
    app.run()
```
