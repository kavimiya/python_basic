# Q1. What is Flask Framework? What are the advantages of Flask Framework?


Flask is a lightweight Python web framework for building web applications with simplicity and flexibility.

Flask's advantages are simplicity, flexibility, lightweight, extensibility, great documentation, built-in development server, RESTful support, Jinja2 templating, and production-readiness.

# Q2. Create a simple Flask application to display ‘Hello World!!’. Attach the screenshot of the output in Jupyter Notebook.

```python
# Importing the required modules
from flask import Flask

# Create a Flask application instance
app = Flask(__name__)

# Define a route and a view function
@app.route('/')
def hello_world():
    return 'Hello World!!'

# Run the application
if __name__ == '__main__':
    app.run()
```

# Q3. What is App routing in Flask? Why do we use app routes?

* App routing in Flask is the process of mapping URL paths to view functions that handle incoming requests.
* We use app routes to define different endpoints in our web application and determine the content to serve to users based on the URLs they visit. 
* It helps organize the application's functionality and ensures a clear structure for handling different requests.

# Q4.  Create a “/welcome” route to display the welcome message “Welcome to ABC Corporation” and a “/” route to show the following details:

    Company Name: ABC Corporation

    Location: India

    Contact Detail: 999-999-9999 


Attach the screenshot of the output in Jupyter Notebook.

```python
from flask import Flask

app = Flask(__name__)

# Route for /welcome
@app.route('/welcome')
def welcome():
    return 'Welcome to ABC Corporation'

# Route for /
@app.route('/')
def company_details():
    return '''Company Name: ABC Corporation\n
              Location: India\n
              Contact Detail: 999-999-9999'''

if __name__ == '__main__':
    app.run()
```

# Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the url_for() function

```python
from flask import Flask, url_for, redirect, render_template

app = Flask(__name__)

# Route for the homepage
@app.route('/')
def home():
    return 'Welcome to the homepage!'

# Route for the contact page
@app.route('/contact')
def contact():
    return 'Contact us at contact@example.com'

# Route for a dynamic page that takes a username parameter
@app.route('/user/<username>')
def user_profile(username):
    return f'Welcome to the profile page of {username}'

# Route to demonstrate using url_for()
@app.route('/welcome')
def welcome():
    # Using url_for() to generate URLs for the defined endpoints
    home_url = url_for('home')
    contact_url = url_for('contact')
    user_url = url_for('user_profile', username='JohnDoe')

    # Rendering the template with the generated URLs
    return render_template('welcome.html', home_url=home_url, contact_url=contact_url, user_url=user_url)

if __name__ == '__main__':
    app.run()
```
